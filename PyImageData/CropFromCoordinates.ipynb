```python
#Load Libraries
import pandas as pd
import fnmatch
import glob
import os
import glob
from PIL import Image
```

```python
#find and load all of the curbramp.txt files
AllCurbRamp=pd.DataFrame()
list_=[]

for filename in glob.iglob('DataSample/**/*_curbramp.txt', recursive=True):
    #print(filename)
    temp=pd.read_table(filename, header=None, sep=" ")
    temp['filename']=filename
    list_.append(temp)
AllCurbRamp=pd.concat(list_)
AllCurbRamp.columns=[['x1', 'x2', 'y1', 'y2', 'filename']]

#extract the image name
AllCurbRamp['imagename']=AllCurbRamp['filename'].map(lambda x: str(x)[:-13])
AllCurbRamp['imagename']=AllCurbRamp['imagename'].map(lambda x: str(x)[14:])

#Add Indicator for CurbRamp so when we merge with non-curb ramp data later it will output for classification
AllCurbRamp['CurbRamp']=1

AllCurbRamp.head()
```

```python
#find and load all of the nocurbramp.txt files

NoCurbRamp=pd.DataFrame()
list_=[]

for filename in glob.iglob('DataSample/**/*_nocurbramp.txt', recursive=True):
    #print(filename)
    temp=pd.read_table(filename, header=None, sep=" ")
    temp['filename']=filename
    list_.append(temp)
NoCurbRamp=pd.concat(list_)
NoCurbRamp.columns=[['x1', 'x2', 'y1', 'y2', 'filename']]

#extract imagename
NoCurbRamp['imagename']=NoCurbRamp['filename'].map(lambda x: str(x)[:-15])
NoCurbRamp['imagename']=NoCurbRamp['imagename'].map(lambda x: str(x)[14:])

#AddIndicator for CurbRamp so when we merge with non-curb ramp data later it will be a an output for classification
NoCurbRamp['CurbRamp']=1


NoCurbRamp.head()
```

```python
NewDir="NoCurb/"

#Outer Loop: Goes through images
for infile in glob.iglob("DataSample/**/*.jpg", recursive=True):
    #infile="DataSample/-2/-2q0r9oa712WtzJ4J9WjBw.jpg"
    file, ext = os.path.splitext(infile)
    CurrDir, file =os.path.split(file)


#Get all of the NoCurbRamps Crop Boxes
    ThisImageCoord=pd.DataFrame(NoCurbRamp.loc[NoCurbRamp['imagename']==file])


    for index, row in ThisImageCoord.iterrows():
        #define a tuple for the bounding box
        boundaries=(row[0], row[2], row[1], row[3])
        #open the image
        im = Image.open(infile)
        #Crop it based on the boundaries
        im = im.crop(boundaries)
        #Create a sequence name for this image
        crop=str(row.name)
        #Full file name and save it out to the new directory
        savefile=NewDir+file + "-" + crop + ".jpg"
        im.save(savefile)
```
