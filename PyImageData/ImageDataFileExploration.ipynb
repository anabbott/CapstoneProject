```python
#XML to Pandas
#http://www.austintaylor.io/lxml/python/pandas/xml/dataframe/2016/07/08/convert-xml-to-pandas-dataframe/

import xml.etree.ElementTree as ET
import pandas as pd

#xml_data = open('DataSample/1a/1A00FcZyk-UiLb7KtLBubQ.xml').read()

def xml2df(xml_data):
    root = ET.XML(xml_data) # element tree
    all_records = []
    for i, child in enumerate(root):
        record = {}
        for subchild in child:
            record[subchild.tag] = subchild.text
            all_records.append(record)
    return pd.DataFrame(all_records)
```

```python
AllXML=pd.DataFrame()
list_=[]

for filename in glob.iglob('DataSample/**/*.xml', recursive=True):
    #print(filename)
    xml_data = open(filename).read()
    temp=xml2df(xml_data)
    temp['filename']=filename
    list_.append(temp)
AllXML=pd.concat(list_)

```

```python
#extract the image name
AllXML['imagename']=AllXML['filename'].map(lambda x: str(x)[:-4])
AllXML['imagename']=AllXML['imagename'].map(lambda x: str(x)[14:])



AllXML.head()
```
